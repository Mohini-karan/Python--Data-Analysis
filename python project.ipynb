```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# drop columns
df.drop(["Status", "unnamed1"], axis = 1, inplace = True)
```

```python
df['Amount'] = df['Amount'].replace(np.nan, 9453.61 )
```

```python
df["Amount"] = df["Amount"]. astype("int")
```

```python
df["Amount"].dtypes
```

```python
df[["Age", "Orders", "Amount"]].describe()
```

```python
#Exploratory Data Analysis
#Gender
plt = sns.countplot(x = "Gender", data= df, palette= "viridis")
for bars in plt.containers:
    plt.bar_label(bars)
```

```python
x = df.groupby(["Gender"], as_index= False).agg({"Amount": "sum"}).sort_values(by="Amount", ascending= False)
print(x)
```

```python
sns.barplot(x= "Gender", y= "Amount", data= x,palette= "spring" )
```

```python
#From above graphs we can see that most of the buyers are femals and even purchasing power of females are greater than men
```

```python
#Age
y = sns.countplot(x= "Age Group", hue= "Gender", data= df)
for bars in y.containers:
    y.bar_label(bars)
```

```python
sales_age = df.groupby(["Age Group"], as_index= False).agg({"Amount": "sum"}).sort_values(by="Amount", ascending= False)
sns.barplot(x= "Age Group", y = "Amount", data = sales_age, palette= "viridis")
```

```python
#From above graphs we can see that most of the buyers are of age group between 26-35 yrs female
```

```python
#state
sales_state = df.groupby(["State"], as_index= False).agg({"Orders": "sum"}).sort_values(by= "Orders", ascending= False).head(10)
sns.set(rc= {"figure.figsize": (15,5)})
sns.barplot(data = sales_state,x= "State", y= "Orders", palette= "deep")
```

```python
sales_state = df.groupby(["State"], as_index= False).agg({"Amount": "sum"}).sort_values(by= "Amount", ascending= False).head(10)
sns.set(rc= {"figure.figsize": (15,5)})
sns.barplot(data = sales_state,x= "State", y= "Amount", palette= "bright")
```

```python
#from above graphs we can see that most of the orders & total sales/ amount are from uttar Pradesh, Maharashtra and Karnataka
```

```python
#Maritial status
marital_status= sns.countplot(data= df, x = "Marital_Status", palette= "pastel")
sns.set(rc={"figure.figsize": (7,5)})
for bars in marital_status.containers:
    marital_status.bar_label(bars)
```

```python
sales_state = df.groupby(["Marital_Status", "Gender"], as_index= False).agg({"Amount": "sum"}).sort_values(by= "Amount", ascending= False)
sns.set(rc= {"figure.figsize": (6,5)})
sns.barplot(data = sales_state,x= "Marital_Status", y= "Amount",hue= "Gender", palette= "colorblind")
```

```python
# from above graphs we can see that most of the buyers are married(women) and they have high purchasing power
```

```python
sns.set(rc={"figure.figsize": (20,7)})
ax = sns.countplot(x = "Occupation", data= df, palette= "deep")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(["Occupation"], as_index= False).agg({"Amount": "sum"}).sort_values(by= "Amount", ascending= False)
sns.barplot(data= sales_state, x= "Occupation", y= "Amount", palette= "deep")
```

```python
# from above graph we can see that most of the buyers are working in IT, Healthcare, Aviation sector     
```

```python
#product categeory
sns.set(rc={"figure.figsize": (25,5)})
ax = sns.countplot(x = "Product_Category", data= df, palette= "deep")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(["Product_Category"], as_index= False).agg({"Amount": "sum"}).sort_values(by= "Amount", ascending= False).head(10)
sns.set(rc={"figure.figsize": (15,5)})
sns.barplot(data= sales_state, x= "Product_Category", y= "Amount", palette= "viridis")
```

```python
# from above graphs we can see that most of the sold product  are from Food, Clothing, and Electronics categeory
```

```python
#conclusion
#Married women age group  26-35 yrs from up, maharashtra and karnataka working in it , healthcare, and aviation are more likely buy product from food, Clothing and Electronics categeory
```
